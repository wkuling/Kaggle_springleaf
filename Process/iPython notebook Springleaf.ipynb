```python
# Helper function to process some of the day observations

def timetodaysago(timeobs):
    try:
        d0 = datetime.strptime(timeobs[:7], '%d%b%y')
        d1 = datetime.strptime('01SEP15', '%d%b%y')
        delta = d1 - d0
        return delta.days
    except:
        return timeobs
```

```python
def preprocess(block, block2):
    
# Preprocessing of Pandas dataframe
# block = train, block2 = test

# Walk through DateFrame column by column
# 1) Drop column if only 1 value is there (useless)
# 2) Transform timestamps to a continuous variable
# 3) Just a continuous variable with cardinality > 12 (keep as is)
# 4) Cardinality <= 60 (either value or object)
# 5) Cardinality > 60 (truncated one-hot encoding)
    
    for ii in block.columns:
        print "Checking: " + str(ii)
        print
        
        top_keys = pd.Series(list(block[ii].value_counts().keys()))
        top_values = pd.Series(list(block[ii].value_counts()))
        to_check_series = top_keys[top_keys<>-9999]
        to_check = list(to_check_series)[0] # Ugly, but necessary to get index right
        
        print top_keys[:10]
        print
        print "To check: " + str(to_check)
        raw_cardinality = len(top_keys.value_counts())
        print "Raw Cardinality: " + str(raw_cardinality)
        
        if raw_cardinality <= 1: #So there is a only one value in this column
            print " Only one or zero distinct values --> drop"
            block.drop(ii, axis = 1, inplace = True)
            block2.drop(ii, axis = 1, inplace = True)
        else: # Checking whether we have situation 2 or 3/4/5 now
            print " Checking for datetimeformat: " + str(to_check)
            p = re.compile(r'\w{7}:\d{2}:\d{2}:\d{2}') # regex for our datetime format
            if p.match(str(to_check)): # Situation 2: match
                print " Timestamp format --> mapping to timetodaysago"
                block[ii]=block[ii].map(timetodaysago)
                block2[ii]=block2[ii].map(timetodaysago)
                block[ii]=block[ii].astype('float64')
                block2[ii]=block2[ii].astype('float64')
            if block[ii].dtypes in [np.float64, np.int64, np.uint16, np.uint32, np.uint8] and raw_cardinality>112:
                # Situation 3: regular value type
                print " All types are of regular value type, raw_cardinality > 12... pass and keep the same" 
                pass
            else:
                # Situation 4 and 5: bin the top-60 variables
                top_keys = top_keys[top_values>=30][:60]
                cardinality_corrected = len(top_keys)
                print " Cardinality corrected for at least 30 observations: " + str(cardinality_corrected)
                if cardinality_corrected > 1:
                    block[ii] = block[ii].map(lambda x: x if (x in list(top_keys)) else ('_other'))
                    block2[ii] = block2[ii].map(lambda x: x if (x in list(top_keys)) else ('_other'))
                    toadd = pd.get_dummies(block[ii], prefix = ii)
                    toadd = toadd.astype(np.uint8)
                    toadd.drop(toadd.columns[[len(toadd.columns)-1]], axis = 1, inplace = True)
                    block = pd.concat([block, toadd], axis=1)
                    
                    toadd = pd.get_dummies(block2[ii], prefix = ii)
                    toadd = toadd.astype(np.uint8)
                    toadd.drop(toadd.columns[[len(toadd.columns)-1]], axis = 1, inplace = True)
                    block2 = pd.concat([block2, toadd], axis=1)   
                block.drop(ii, axis = 1, inplace = True)
                block2.drop(ii, axis = 1, inplace = True)

    return block, block2
                        
                    
    
```

```python
train, test = preprocess(train, test)
```
