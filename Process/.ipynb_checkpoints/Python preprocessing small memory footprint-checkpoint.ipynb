```python
def timetodaysago (timeobs):
    # input: timeobs is of format 23NOV10:00:00:00
    # output: days ago since September 1st, 2015
    
    try:
        d0= datetime.strptime(timeobs[:7], '%d%b%y')
        d1= datetime.strptime('01SEP15', '%d%b%y')
        delta = d1 - d0
        return delta.days
    except:
        return timeobs

# Test by removing # in line below
# timetodaysago('04AUG12:00:00:00') 
```

```python
def preprocess(block):
    
# Processing the columns of the dataframe:
# Input: block = Pandas Dataframe
# Output: Pandas Dataframe which represents Pandas Series

# Now, there are five situations per column:
# 1) Drop columns if only 1 value is there (useless) (note that some values and some NAs are not immediately useless)
# 2) Transform timestamps to a continuous variable
# 3) Check whether all instances are 'values' -> cast them to float 64
# 4) Make dummies for columns with 1 < cardinality < 60
# 5) Cardinality >= 60 and 'names' -> Dummies for most frequent variables

    for ii in block.columns:
        print 'Checking: ' + str(ii)
        print

        top_keys = pd.Series(list(block[ii].value_counts().keys()))
        top_values = pd.Series(list(block[ii].value_counts()))    
        to_check_series = top_keys[top_keys<>-9999]
        to_check = list(to_check_series)[0] # This ugly trick is necessary as the index doesn't start at 0 for a selected DF :(

        print top_keys[:10]
        print   
        print 'To check: ' + str(to_check)
        raw_cardinality = len(top_keys.value_counts()) # Note that although a variable might be only False, NA then also contains info
        print 'Raw Cardinality: ' + str(raw_cardinality)

        if raw_cardinality <= 1: # So there is max a single value in all columns (situation 1) --> drop!
            print 'Only one or zero distinct values --> drop'
            block.drop(ii, axis=1, inplace=True)
        else: # Checking whether we have situation 2 or 3/4/5 now...

            print 'Checking for datetimeformat: ' + str(to_check) # Note that raw cardinality > 1, so there has to be a non-null value

            p = re.compile(r'\w{7}:\d{2}:\d{2}:\d{2}') # Regular expression for datetime format used

            if p.match(str(to_check)): # Situation 2: the match returns something > it's of a datetime format 
                print 'Timestamp format --> mapping to timetodaysago'
                block[ii]=block[ii].map(timetodaysago)
                block[ii]=block[ii].astype('float64')                
            else:
                if block[ii].dtypes in [np.float64, np.int64, np.uint16, np.uint32, np.uint8] and raw_cardinality>12: 
                    # Situation 3: all real values
                    print 'All types are of regular value type, raw_cardinality>12... pass and leave untouched'
                    pass            
                else: # Checking whether we have situation 4 or 5 now...                           
                    if raw_cardinality < 60: # This would be situation 4 --> Let's one hot encode this!
                        print 'Making ' + str(raw_cardinality) + ' dummies'
                        toadd = pd.get_dummies(block[ii], prefix=ii) # Generate dummies
                        toadd = toadd.astype(np.uint8) # Cast to smaller memory footprint datatype
                        # Drop the last column generated, avoiding the dummy variable trap
                        toadd.drop(toadd.columns[[len(toadd.columns)-1]], axis=1, inplace=True) 
                        block = pd.concat([block, toadd], axis=1) # Add the dummies to the DataFrame
                        # Dropping the original column, as it has been one hot encoded now.
                        block.drop(ii, axis=1, inplace=True) 
                    else: # So, situation 5 with large cardinality and few observations (TODO: binning!!)
                        top_keys = top_keys[top_values>=30][:60] # So, observations with n>=30 with max 60 gets a seperate dummy                   
                        cardinality_corrected=len(top_keys)
                        print 'Cardinality corrected for at least 30 observations: ' + str(cardinality_corrected)
                        if cardinality_corrected >= 1:
                            print 'Making ' + str(cardinality_corrected + 1) + ' dummies'
                            block[ii]=block[ii].map(lambda x: x if (x in list(top_keys)) else ('_other')) 
                            # And maps the values to this top (or var_other)
                            toadd = pd.get_dummies(block[ii], prefix=ii) 
                            # Generate dummies (for variables with n > 30)
                            toadd = toadd.astype(np.uint8) # Cast to smaller memory footprint datatype
                            toadd.drop(toadd.columns[[len(toadd.columns)-1]], axis=1, inplace=True) 
                            # Drop the last column generated, avoiding the dummy variable trap
                            block = pd.concat([block, toadd], axis=1) # Add the dummies to the DataFrame
                        block.drop(ii, axis=1, inplace=True) # Dropping the original column, 
                                                             # as it has been one hot encoded now (or it's too sparse)
    return block
```
